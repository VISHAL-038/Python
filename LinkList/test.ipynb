```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def add_first(self, data):
        new_node = Node(data)
        self.size += 1
        if self.head is None:
            self.head = self.tail = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node

    def add_last(self, data):
        new_node = Node(data)
        self.size += 1
        if self.head is None:
            self.head = self.tail = new_node
            return
        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node

    def add_middle(self, idx, data):
        if idx == 0:
            self.add_first(data)
            return
        new_node = Node(data)
        self.size += 1
        temp = self.head
        i = 0
        while i < idx - 1:
            temp = temp.next
            i += 1
        new_node.next = temp.next
        if temp.next:
            temp.next.prev = new_node
        temp.next = new_node
        new_node.prev = temp
        if new_node.next is None:
            self.tail = new_node

    def remove_first(self):
        if self.head is None:
            return
        self.size -= 1
        temp = self.head
        self.head = self.head.next
        if self.head is None:  
            self.tail = None
        else:
            self.head.prev = None
        temp.next = None

    def remove_last(self):
        if self.head is None:
            return
        self.size -= 1
        if self.head.next is None:  
            self.head = self.tail = None
            return
        temp = self.tail
        self.tail = self.tail.prev
        self.tail.next = None
        temp.prev = None

    def print_list(self):
        if self.head is None:
            print("DoublyLinkedList is empty")
            return
        temp = self.head
        while temp is not None:
            print(f"{temp.data}<->", end="")
            temp = temp.next
        print("end")

if __name__ == "__main__":
    dll = DoublyLinkedList()
    dll.add_first(1)
    dll.add_first(2)
    dll.add_last(3)
    dll.add_middle(1, 4)
    dll.print_list()
    dll.remove_first()
    dll.print_list()
    dll.remove_last()
    dll.print_list()
```
