```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def kth_level_elements(root, k):
    if not root:
        return []

    queue = [(root, 0)]  # (node, level)
    result = []

    while queue:
        node, level = queue.pop(0)
        
        if level == k:
            result.append(node.value)
        
        if node.left:
            queue.append((node.left, level + 1))
        
        if node.right:
            queue.append((node.right, level + 1))

    return result

# Example usage
# Constructing the binary tree
#        1
#       / \
#      2   3
#     / \ / \
#    4  5 6  7

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

k = 2
print(f"Elements at level {k}: {kth_level_elements(root, k)}")
```

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def find_LCA(root, p, q):
    if not root:
        return None
    
    if root == p or root == q:
        return root
    
    left = find_LCA(root.left, p, q)
    right = find_LCA(root.right, p, q)
    
    if left and right:
        return root
    
    return left if left else right

# Example usage
# Constructing the binary tree
#        3
#       / \
#      5   1
#     / \ / \
#    6  2 0  8
#      / \
#     7   4

root = TreeNode(3)
root.left = TreeNode(5)
root.right = TreeNode(1)
root.left.left = TreeNode(6)
root.left.right = TreeNode(2)
root.right.left = TreeNode(0)
root.right.right = TreeNode(8)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

p = root.left  # Node with value 5
q = root.left.right.right  # Node with value 4

lca = find_LCA(root, p, q)
if lca:
    print(f"The LCA of {p.value} and {q.value} is {lca.value}")
else:
    print("LCA not found.")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, data):
    if root is None:
        return Node(data)
    else:
        if data < root.data:
            root.left = insert(root.left, data)
        else:
            root.right = insert(root.right, data)
    return root

def binaryTreePaths(root):
    def dfs(node, path, paths):
        if not node:
            return
        
        # Append the current node value to the path
        if path:
            path += "->" + str(node.data)
        else:
            path = str(node.data)
            
        # If it's a leaf node, add the path to the result
        if not node.left and not node.right:
            paths.append(path)
            return
        
        # Recursively call dfs for left and right children
        if node.left:
            dfs(node.left, path, paths)
        if node.right:
            dfs(node.right, path, paths)
    
    # List to store all the paths
    paths = []
    dfs(root, "", paths)
    return paths

def inorderTraversal(root):
    result = []
    def inorder(node):
        if not node:
            return
        inorder(node.left)
        result.append(node.data)
        inorder(node.right)
    inorder(root)
    return result

# Example usage
# Creating a BST [4, 3, 6, 1, 2, 5, 7]
root = None
elements = [4, 3, 6, 1, 2, 5, 7]
for element in elements:
    root = insert(root, element)

# Get all root-to-leaf paths
print(binaryTreePaths(root))  # Output: ["4->3->1->2", "4->6->5", "4->6->7"]

# Get inorder traversal
print(inorderTraversal(root))  # Output: [1, 2, 3, 4, 5, 6, 7]
```

```python
# print in range BST?
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def rangeBST(root, low, high):
    result = []
    
    def inorder(node):
        if node:
            # Traverse the left subtree
            inorder(node.left)
            
            # Check if the current node's value is within the range
            if low <= node.val <= high:
                result.append(node.val)
            
            # Traverse the right subtree
            inorder(node.right)
    
    inorder(root)
    return result

# Example usage
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(15)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.left = TreeNode(13)
root.right.right = TreeNode(18)

low = 6
high = 14

print(rangeBST(root, low, high))  
```

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def rootToLeafPaths(root):
    paths = []
    
    def findPaths(node, current_path):
        if node:
            # Add the current node to the path
            current_path.append(node.val)
            
            # If it's a leaf node, add the path to the result
            if not node.left and not node.right:
                paths.append(list(current_path))
            else:
                # Otherwise, continue to explore the children
                findPaths(node.left, current_path)
                findPaths(node.right, current_path)
            
            # Backtrack: remove the current node from the path
            current_path.pop()
    
    findPaths(root, [])
    return paths

# Example usage
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(15)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.left = TreeNode(13)
root.right.right = TreeNode(18)

print(rootToLeafPaths(root))  # Output: [[10, 5, 3], [10, 5, 7], [10, 15, 13], [10, 15, 18]]
```

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def is_valid_bst(node, left=float('-inf'), right=float('inf')):
    if node is None:
        return True

    if not (left < node.value < right):
        return False

    return (is_valid_bst(node.left, left, node.value) and
            is_valid_bst(node.right, node.value, right))

# Example usage:
# Constructing a valid BST
#       5
#      / \
#     3   7
#    / \   \
#   2   4   8

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)

print(is_valid_bst(root))  # Output: True
```

```python
# Node class definition for a binary tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Node class definition for a doubly linked list
class DLLNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

# Helper function to convert binary tree to doubly linked list
def convert_to_dll(root):
    if not root:
        return None

    # Helper variables to keep track of the previous and head nodes
    prev = None
    head = None

    # In-order traversal to process the nodes
    def inorder_traversal(node):
        nonlocal prev, head
        if not node:
            return

        # Recursively traverse the left subtree
        inorder_traversal(node.left)

        # Convert the TreeNode to DLLNode
        current = DLLNode(node.value)

        # If prev is None, it means this is the leftmost node (head of DLL)
        if prev is None:
            head = current
        else:
            current.prev = prev
            prev.next = current

        # Update prev to current
        prev = current

        # Recursively traverse the right subtree
        inorder_traversal(node.right)

    # Start in-order traversal from root
    inorder_traversal(root)

    return head

# Helper function to print the doubly linked list
def print_dll(head):
    while head:
        print(head.value, end=" <-> " if head.next else "\n")
        head = head.next

# Example usage
if __name__ == "__main__":
    # Create a binary tree
    root = TreeNode(10)
    root.left = TreeNode(6)
    root.right = TreeNode(14)
    root.left.left = TreeNode(4)
    root.left.right = TreeNode(8)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(16)

    # Convert the binary tree to doubly linked list
    dll_head = convert_to_dll(root)

    # Print the doubly linked list
    print_dll(dll_head)
```

```python
def knapsack(values, weights, W):
    n = len(values)
    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w - weights[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][W]

values = [60, 100, 120]
weights = [10, 20, 30]
W = 50

print("Maximum total value in knapsack:", knapsack(values, weights, W))
```
