```python
# k^th level element

class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def kth_level_elements(root, k):
    if root is None:
        return []

    queue = [(root, 0)] 
    result = []

    while queue:
        node, level = queue.pop(0)
        
        if level == k:
            result.append(node.value)
        
        if node.left:
            queue.append((node.left, level + 1))
        
        if node.right:
            queue.append((node.right, level + 1))

    return result

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

k = 2
kth_level_elements(root, k)
```

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def find_LCA(root, p, q):
    if not root:
        return None
    
    if root == p or root == q:
        return root
    
    left = find_LCA(root.left, p, q)
    right = find_LCA(root.right, p, q)
    
    if left and right:
        return root
    
    return left if left else right


root = TreeNode(3)
root.left = TreeNode(5)
root.right = TreeNode(1)
root.left.left = TreeNode(6)
root.left.right = TreeNode(2)
root.right.left = TreeNode(0)
root.right.right = TreeNode(8)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

p = root.left  
q = root.left.right.right

lca = find_LCA(root, p, q)
if lca:
    print(p.value, q.value)
    print(lca.value)
else:
    print("LCA not found.")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Output:
    def __init__(self):
        self.found = False
        self.lca = None

def lca(root, x, y):
    out = Output()
    if root is None:
        return out
    
    left = lca(root.left, x, y)
    right = lca(root.right, x, y)

    if left.found and right.found:
        out.found = True
        out.lca = root
    elif (root.data == x or root.data == y) and (left.found or right.found):
        out.found = True
        out.lca = root
    elif root.data == x or root.data == y:
        out.found = True
    elif left.found:
        out = left
    elif right.found:
        out = right
    
    return out

def LCA(root, left, right):
    return lca(root, left.data, right.data).lca

if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)

    lca_node = LCA(root, root.left.left, root.left.right)
    print(lca_node.data)
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def binaryTreePaths(self, root):
        """
        :type root: TreeNode
        :rtype: List[str]
        """
        def dfs(node, path, paths):
            if not node:
                return
            
            # Append the current node value to the path
            if path:
                path += "->" + str(node.val)
            else:
                path = str(node.val)
                
            # If it's a leaf node, add the path to the result
            if not node.left and not node.right:
                paths.append(path)
                return
            
            # Recursively call dfs for left and right children
            if node.left:
                dfs(node.left, path, paths)
            if node.right:
                dfs(node.right, path, paths)
        
        # List to store all the paths
        paths = []
        dfs(root, "", paths)
        return paths

# Example usage
# Creating a BST [1,2,3,null,5]
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.right = TreeNode(5)

sol = Solution()
print(sol.binaryTreePaths(root))  # Output: ["1->2->5", "1->3"]
```

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
    
def rangeBST(root, low, high):
    result = []
    
    def inorder(node):
        if node:
            #left 
            inorder(node.left)

            #check if the value is in range
            if low <= node.val <= high:
                result.append(node.val)

            #right
            inorder(node.right)

    inorder(root)
    return result
    
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(15)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.left = TreeNode(13)
root.right.right = TreeNode(18)

low = 1
high = 13

print(rangeBST(root, low, high))  
```

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def rootToLeafPaths(root):
    paths = []
    def findPaths(node, current_path):
        if node:
            current_path.append(node.val)

            if not node.left and not node.right:
                paths.append(list(current_path))
            else:
                findPaths(node.left, current_path)
                findPaths(node.right, current_path)

            current_path.pop()
    
    findPaths(root, [])
    return paths


root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(15)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.left = TreeNode(13)
root.right.right = TreeNode(18)

print(rootToLeafPaths(root))
```

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def mirrorBST(root):
    if root is None:
        return
   
    # Swap the left and right children
    root.left, root.right = root.right, root.left
    
    # Recursively call mirrorBST on left and right subtrees
    mirrorBST(root.left)
    mirrorBST(root.right)


def inorderTraversal(root):
    if root:
        inorderTraversal(root.left)
        print(root.val, end=' ')
        inorderTraversal(root.right)

# Constructing the BST
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(15)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.left = TreeNode(13)
root.right.right = TreeNode(18)

print("Original BST (in-order):")
inorderTraversal(root)
print("\n")

# Mirroring the BST
mirrorBST(root)

print("Mirrored BST (in-order):")
inorderTraversal(root)
print("\n")
```

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def sortedArrayToBST(nums):
    if not nums:
        return None

    # Find the middle element
    mid = len(nums) // 2

    # Create the root node with the middle element
    root = TreeNode(nums[mid])

    # Recursively build the left subtree
    root.left = sortedArrayToBST(nums[:mid])

    # Recursively build the right subtree
    root.right = sortedArrayToBST(nums[mid+1:])

    return root

def inorderTraversal(root):
    if root:
        inorderTraversal(root.left)
        print(root.val, end=' ')
        inorderTraversal(root.right)

# Sorted array
nums = [-10, -3, 0, 5, 9]

# Convert sorted array to BST
bst_root = sortedArrayToBST(nums)

# Print the BST in-order
print("In-order Traversal of the BST:")
inorderTraversal(bst_root)
print("\n")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def sortedArrayToBST(self, arr, start, end):
        if start > end:
            return None
        mid = (start + end) // 2
        root = Node(arr[mid])
        root.left = self.sortedArrayToBST(arr, start, mid - 1)
        root.right = self.sortedArrayToBST(arr, mid + 1, end)
        return root

    def inorder(self, root):
        if root is None:
            return
        self.inorder(root.left)
        print(root.data, end=" ")
        self.inorder(root.right)

    def preorder(self, root):
        if root is None:
            return
        print(root.data, end=" ")
        self.preorder(root.left)
        self.preorder(root.right)

    def postorder(self, root):
        if root is None:
            return
        self.postorder(root.left)
        self.postorder(root.right)
        print(root.data, end=" ")

if __name__ == "__main__":
    arr = [1, 2, 3, 4, 5, 6, 7]
    bst = BST()
    root = bst.sortedArrayToBST(arr, 0, len(arr) - 1)
    print("Inorder: ", end="")
    bst.inorder(root)
    print()
    print("Preorder: ", end="")
    bst.preorder(root)
    print()
    print("Postorder: ", end="")
    bst.postorder(root)
```

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def is_valid_bst(node, left=float('-inf'), right=float('inf')):
    if node is None:
        return True

    if not (left < node.value < right):
        return False

    return (is_valid_bst(node.left, left, node.value) and
            is_valid_bst(node.right, node.value, right))

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)

print(is_valid_bst(root))
```

Binary Tree Converted into Doubly link list

```python
class TreeNode:
    def __init__(self, item):
        self.data = item
        self.left = None
        self.right = None

class BinaryTreeToDll:
    def __init__(self):
        self.root = None
        self.head = None
        self.prev = None

    def treeToDll(self, node):
        if node is None:
            return

        self.treeToDll(node.left)

        if self.prev is None:
            self.head = node
        else:
            node.left = self.prev
            self.prev.right = node  
        self.prev = node

        self.treeToDll(node.right)

    def printDll(self, node):
        while node is not None:
            print(node.data, end=" ")
            node = node.right

if __name__ == "__main__":
    tree = BinaryTreeToDll()
    tree.root = TreeNode(10)
    tree.root.left = TreeNode(12)
    tree.root.right = TreeNode(15)
    tree.root.left.left = TreeNode(25)
    tree.root.left.right = TreeNode(30)
    tree.root.right.left = TreeNode(36)

    tree.treeToDll(tree.root)

    tree.printDll(tree.head)
```
